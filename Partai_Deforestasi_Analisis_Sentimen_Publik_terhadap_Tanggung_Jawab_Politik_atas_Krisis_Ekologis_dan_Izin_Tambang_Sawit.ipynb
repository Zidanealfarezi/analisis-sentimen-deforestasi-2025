```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re

# Load dataset dari direktori Colab
file_path = '/content/tweets-data/penggundulanhutan.csv'
df = pd.read_csv(file_path)

print(f"Data berhasil dimuat: {len(df)} baris.")
df.head()
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|@\w+|#\w+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_text'] = df['full_text'].apply(clean_text)

# Daftar Stopwords untuk membersihkan WordCloud
indo_stopwords = {
    'yang', 'dan', 'di', 'ke', 'dari', 'ini', 'itu', 'dengan', 'ada', 'adalah',
    'untuk', 'pada', 'saya', 'kita', 'ia', 'bahwa', 'tidak', 'tapi', 'namun',
    'juga', 'karena', 'sudah', 'telah', 'saat', 'hanya', 'bisa', 'tak', 'jadi',
    'oleh', 'saja', 'dalam', 'amp', 'ingin', 'buat', 'kalau', 'paling', 'agak',
    'lah', 'kok', 'aja', 'ya', 'ni', 'sih'
}

print("Pembersihan teks selesai.")
```

```python
plt.figure(figsize=(12, 6))
all_words = ' '.join(df['cleaned_text'])
wordcloud = WordCloud(width=1000, height=500,
                      background_color='white',
                      stopwords=indo_stopwords,
                      colormap='OrRd',
                      max_words=60).generate(all_words)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Peta Percakapan: Deforestasi & Krisis Ekologis 2025', fontsize=18, pad=20, fontweight='bold')
plt.show()
```

```python
neg_words = [
    'rusak', 'bencana', 'gagal', 'parah', 'krisis', 'ironi', 'pencitraan',
    'banjir', 'liar', 'salah', 'merusak', 'abaikan', 'serakah', 'eksploitasi',
    'masif', 'cuci tangan', 'partai deforestasi'
]
pos_words = ['baik', 'dukung', 'solusi', 'peduli', 'reboisasi', 'lestari', 'hijau', 'aman', 'lindungi']

def get_sentiment(text):
    text = str(text).lower()
    score = 0
    if 'partai deforestasi' in text: score -= 2
    for word in neg_words:
        if word in text: score -= 1
    for word in pos_words:
        if word in text: score += 1
    return 'Negative' if score < 0 else ('Positive' if score > 0 else 'Neutral')

df['sentiment'] = df['cleaned_text'].apply(get_sentiment)

# Visualisasi Distribusi Sentimen
plt.figure(figsize=(9, 6))
colors = {'Negative': '#d63031', 'Neutral': '#636e72', 'Positive': '#27ae60'}
ax = sns.countplot(data=df, x='sentiment', palette=colors, order=['Negative', 'Neutral', 'Positive'])

total = len(df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    ax.annotate(f'{int(p.get_height())} ({percentage})', (p.get_x() + 0.15, p.get_height() + 5), fontweight='bold')

plt.title('Sentimen Publik: Perlawanan terhadap "Partai Deforestasi"', fontsize=15, fontweight='bold')
plt.show()
```

```python
entities = {
    'AHY': df['cleaned_text'].str.contains('ahy').sum(),
    'SBY': df['cleaned_text'].str.contains('sby').sum(),
    'Prabowo': df['cleaned_text'].str.contains('prabowo').sum(),
    'Menhut': df['cleaned_text'].str.contains('menhut').sum(),
    'Mendagri': df['cleaned_text'].str.contains('mendagri').sum()
}
entity_df = pd.DataFrame(list(entities.items()), columns=['Tokoh', 'Sebutan']).sort_values('Sebutan', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(data=entity_df, x='Sebutan', y='Tokoh', palette='flare')
plt.title('Aktor Politik dalam Diskursus Deforestasi', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
